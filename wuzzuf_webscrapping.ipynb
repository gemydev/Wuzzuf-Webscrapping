```python
job_titles = soup.find_all("h2", {"class":"css-m604qf",})  
company_names = soup.find_all("a", {"class":"css-17s97q8",}) 
location_names = soup.find_all("span", {"class":"css-5wys0k",})
job_skills = soup.find_all("div",{'class':"css-y4udm8"})
posted_since = soup.find_all("div",{ 'class':["css-do6t5g" , "css-4c4ojb"] ,})
```

```python
links_of_titles = []
salaries = []
```

```python
list_of_job_titles = []
list_of_company_names = [] 
list_of_location_names = []
list_of_job_skills = []
list_of_posted_since = []
```

```python
for i in range(len(job_titles)):
    list_of_job_titles.append(job_titles[i].text)
    list_of_company_names.append(company_names[i].text)
    list_of_location_names.append(location_names[i].text)
    list_of_job_skills.append(job_skills[i].text)
    list_of_posted_since.append(posted_since[i].text)
    links_of_titles.append(job_titles[i].find("a").attrs["href"])
```

```python
links_of_titles
```

```python
for link in links_of_titles:
    result = requests.get("https://wuzzuf.net"+link)
    src = result.content
    soup = BeautifulSoup(src , "lxml")
    salaries = soup.find_all("span" , {"class":"css-4xky9y"})
```

```python
salaries
```

```python
file_list = [list_of_job_titles , list_of_company_names ,
             list_of_location_names ,list_of_posted_since ,
            list_of_job_skills]

exported = zip_longest(*file_list)
```

```python
with open("wuzzuf_flutter_jobs.csv" , "w") as my_file:
    wr = csv.writer(my_file)
    wr.writerow(["title" , "company" , "location" , "posted_since" , "skills"])
    wr.writerows(exported)
```
